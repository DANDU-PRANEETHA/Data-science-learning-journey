```python
# Load Dataset
import pandas as pd
import numpy as np

df=pd.read_csv("adult_with_headers.csv")
df.info()
```

```python
(df==0).sum()
```

```python
df.isnull().sum()
```

```python
# 1) Handle missing values
# They represent “no gain/loss,” which is real information.
# so keep them as it is
```

```python

cat_cols=['workclass','education','marital_status','occupation',
         'relationship','race','sex','native_country']
```

```python
type(df)
```

```python
# 2) ENCODING TECHNIQUES  (Label Encoding)

from sklearn.preprocessing import LabelEncoder,OneHotEncoder

LE=LabelEncoder()
for col in cat_cols:
    df[col]=LE.fit_transform(df[col].astype(str))

print("Encoding Technique (Label Encoding) ")
df.head()
```

```python
OH=OneHotEncoder(drop='first', sparse_output=False)

df_enc=pd.DataFrame(OH.fit_transform(df[cat_cols]),
                    columns=OH.get_feature_names_out(cat_cols))
df=pd.concat([df.drop(columns=cat_cols),df_enc],axis=1)

print("Encoding Technique (One Hot Encoding) ")
df.head()
```

```python
df.info()
```

```python
#  variables

num_cols=df.select_dtypes(include=['int64','float64']).columns
y=df["income"]
```

```python
# 1a)  STANDARD SCALAR

from sklearn.preprocessing import StandardScaler,MinMaxScaler

ss=StandardScaler()
df[num_cols]=ss.fit_transform(df[num_cols])

print("Standard Scaling : ")
df.head()
```

```python
#  1b)  MIN - MAX SCALAR

mm=MinMaxScaler()
df[num_cols]=mm.fit_transform(df[num_cols])

print("Min - Max Scaling : ")
df.head()
```

```python
'''
StandardScaler
Works best for Linear Regression, Logistic Regression, SVM, PCA etc.,
where models assume normally distributed data

MinMaxScaler
Used in Neural Networks, KNN, Clustering, etc., because they are 
distance-based and sensitive to magnitude
'''
```

```python
#  SUMMARY STATISTICS
df.describe()
```

```python
#  3.    FEATURE ENGINEERING
```

```python
#  "work_hours_category"===Income often depends on how much a person works
#  "capital_diff"===Shows extra income potential

new_cols = pd.DataFrame({
    'work_hours_category': pd.cut(df['hours_per_week'],
                                  bins=[0, 30, 40, 50, 100],
                                  labels=['part-time', 'full-time', 'over-time', 'workaholic']),
    'capital_diff': df['capital_gain'] - df['capital_loss']
})

df = pd.concat([df, new_cols], axis=1).copy()
df.head()
```

```python
# LOG TRANSFORMATION

df['capital_gain_log'] = np.log1p(df['capital_gain'])
df
```

```python
#                                                    Task 2: Dimensionality Reduction with PCA:
```

```python
#    1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.
```

```python
x = data.drop(columns = ["Type"])
y = data["Type"]
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

ss =StandardScaler()
ss_x = pd.DataFrame(ss.fit_transform(x) , columns = x.columns)

print("===================Standard Scaler======================")
ss_x.head()
```

```python
from sklearn.decomposition import PCA

pca =PCA()
pca_x = pd.DataFrame(pca.fit_transform(ss_x))

print("=====================Principal Component Analysis=======================")
pca_x.head()
```

```python
#    2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance) 
#  (or)     cumulative_variance = pca.explained_variance_ratio_.cumsum()

print("Cumulative Explained Variance  : ",cumulative_variance)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()
```

```python
#   Scree plot (Explained Variance per PC)

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(range(1,len(explained_variance)+1),explained_variance, marker = 'o')
plt.title("Scree Plot")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.grid(True)
plt.show()
```

```python
"""
PCA simplifies high-dimensional data.
It reduces noise and improves model efficiency.
Clustering helps group similar data points.
K-Means works well for clearly separated clusters.
DBSCAN is useful for detecting outliers and irregular clusters.
PCA combined with clustering improves overall analysis quality.

"""
```

```python
# Cumulative Explained Variance Plot

plt.figure(figsize=(8,5))
plt.plot(range(1,len(cumulative_variance)+1),cumulative_variance, marker = 'o')
plt.axhline(y=0.95,linestyle = '--')                                                 # 95% reference line
plt.title("Cumulative Explained Variance Plot")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance")
plt.grid(True)
plt.show()
```

```python
#    3.	Transform the original dataset into the principal components.
```

```python
# If cumulative variance reaches 95% at component 6 -> use n_components =6
# If the elbow in scree plot is around PC3 -> use n_components =3  ,here it seems to be 9

pca_final = PCA(n_components = 9)
x_pca_final = pca_final.fit_transform(ss_x)

print("Original Shape :",x.shape)
print("After PCA :",x_pca_final.shape)
```

```python
#                                                      Task 3: Clustering with Original Data:
```

```python
#    1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.
```

```python
from sklearn.cluster import KMeans

n_clusters = 3

model = KMeans(n_clusters=3, random_state = 42 , n_init = 10)
cluster_labels = model.fit_predict(ss_x)
```

```python
#    2.	Visualize the clustering results using appropriate plots.
```

```python
pca_vis = PCA(n_components = 2)
x_orig_2d = pca_vis.fit_transform(ss_x)

plt.scatter(x_orig_2d[:,0] , x_orig_2d[:,1], c = cluster_labels)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Clusters on Original Data")
plt.show()
```

```python
#    3.	Evaluate the clustering performance using metrics such as silhouette score or DaviesBouldin index.
```

```python
from sklearn.metrics import silhouette_score , davies_bouldin_score

sil_orig = silhouette_score(ss_x , cluster_labels)
db_orig = davies_bouldin_score( ss_x , cluster_labels)

print(f"Original data - Silhouette score: {sil_orig:.4f}")
print(f"Original data - Davies - Bouldin index: {db_orig:.4f}")
```

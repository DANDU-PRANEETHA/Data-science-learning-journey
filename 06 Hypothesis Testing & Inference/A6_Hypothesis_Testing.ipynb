```python
#================HYPOTHESIS TESTING==================
```

```python
#===============1. State the Hypotheses statement==============
'''
H0=4000(model predicted and observed are equal)
H1>4000(Observed by owner is high)
'''
```

```python
from scipy import stats
import math
```

```python
#================2. Calculate the Test Statistic===============
```

```python
mu=3050               # owner's observation
mean_units=600
std_units=25
std_dev=5*25
n=25

mu_0=1000+5*mean_units #  model predicted....4000
std_error=std_dev/math.sqrt(n)
```

```python
#===================3. Determine the Probability and compare:===========
```

```python
z_obs = mu - mu_0/std_error   # statistical observed value

alpha=0.05
z_pred=stats.norm.ppf(1-alpha)  # statistical predicted/table value

p_value = stats.norm.sf(z_obs)  # P(Z > z_obs)

print("Model predicted mean (μ₀):", mu_0)
print("Observed sample mean (x̄):", mu)
print("Standard error (SE):", std_error)
print("z_obs =", round(z_obs, 4))
print("z_pred =", round(z_pred, 4))
print("p-value =", round(p_value, 6))


'''Since this is a right-tailed test (we’re checking if costs increased),
a negative z means it’s not in the rejection region.
The p-value will be large, meaning there’s no statistical evidence to prove
costs increased'''
```

```python
#========================4. Make a Decision==========================
```

```python

if z_obs > z_pred:
    print("✅ Owners are right — costs are significantly higher.")
else:
    print("❌ Model is correct — no evidence of higher costs.")
```

```python
#======================5. Conclusion=============================
'''
We fail to reject the null hypothesis (H₀).
There is no significant evidence to support the restaurant
owners’ claim that the weekly operating costs are higher than
predicted by the model.'''
```

```python
from scipy import stats
import numpy as np
import math
```

```python
#  a. Build 99% Confidence Interval Using Sample Standard Deviation
```

```python
d=[1.13, 1.55, 1.43, 0.92,
   1.25, 1.36, 1.32, 0.85, 
   1.07, 1.48, 1.20, 1.33,
   1.18, 1.22, 1.29]
s_mean = np.mean(d)
s_std = np.std(d,ddof=1)
n=15
df=n-1                             # degree of freedom

alpha = 1-0.99
t_val=stats.t.ppf(1-alpha/2,df)    # right region = 1- left region
t_crit = t_val*(s_std/math.sqrt(n))

ci_t = (s_mean - t_crit, s_mean + t_crit)
print(np.round(ci_t,4))
```

```python
#    b. Build 99% Confidence Interval Using Known Population Standard Deviation
```

```python
pop_std = 0.2
z_val=stats.norm.ppf(1-alpha/2)    
z_crit = z_val*(pop_std/math.sqrt(n))

ci_z = (s_mean - z_crit, s_mean + z_crit)
print(np.round(ci_z,4))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('A18 -sonardataset.csv')
data.info()
```

```python
x = data.drop(columns = ['Y'])
y = data['Y']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
ss_x = scaler.fit_transform(x)
```

```python
#                                                            2. Model Implementation
```

```python
#  Divide the dataset into training and test sets.
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(ss_x,y,test_size = 0.2,random_state = 42, stratify = y)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
```

```python
#   Construct a basic ANN model using your chosen high-level neural network library. Ensure your model includes at least one hidden layer.
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input

model_default = Sequential()

model_default.add(Input(shape = (60,)))            # Input Layer
model_default.add(Dense(32, activation = 'relu'))   # Hidden Layer

# Input + Hidden Layer
# model_default.add(Dense(32, input_shape = (60,) , activation = 'relu')

# You may add another hidden layer (optional)
# model_default.add(Dense(16, activation = 'relu'))

# Output Layer ( Binary Classification )
model_default.add(Dense(1, activation = 'sigmoid'))

# Compile model
model_default.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
```

```python
"""
# other optimizers functions
- tf.keras.optimizers.RMSprop()
- tf.keras.optimizers.Adadelta()
- tf.keras.optimizers.Adagrad()
- tf.keras.optimizers.Adamax()
- tf.keras.optimizers.Nadam()
- tf.keras.optimizers.Ftrl()
"""
```

```python
#             Train your model on the training set and then use it to make predictions on the test set.
```

```python
# Fit the model on train data

history = model_default.fit(x_train,
                    y_train_enc,
                    epochs = 50,
                    batch_size = 16,
                    validation_split = 0.2,
                    verbose = 1)
```

```python
'''
VERBOSE

Controls how much training/evaluation info is printed.

verbose=1 â†’ shows progress bar & details
verbose=0 â†’ silent
verbose=2 â†’ one line per epoch
'''
```

```python
# evaluate the model

y_pred_default = model_default.predict(x_test)
y_pred = (y_pred_default > 0.5).astype(int)
```

```python
'''
y_pred = (y_pred_default > 0.5).astype(int)
y_pred_default = prediction probabilities (0â€“1)

> 0.5 â†’ convert to True/False

.astype(int) â†’ convert to 1/0

So:
values â‰¥ 0.5 â†’ class 1
values < 0.5 â†’ class 0
'''
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test_enc, y_pred))
print(classification_report(y_test_enc, y_pred))
```

```python
# Visualize training history

# list all data in history
history.history.keys()

# summarize history for loss

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#                                          3. Hyperparameter Tuning
```

```python
# Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

def create_model(neurons=32, hidden_layers=1, activation='relu', lr=0.001):
    model = Sequential()

    # Input layer + first hidden layer
    model.add(Dense(neurons, input_shape=(60,), activation=activation))

    # Extra hidden layers (if more than one)
    for _ in range(hidden_layers - 1):
        model.add(Dense(neurons, activation=activation))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
#pip install scikeras
```

```python

from scikeras.wrappers import KerasClassifier

model = KerasClassifier(
    model=create_model,
    verbose=0
)
```

```python
from scikeras.wrappers import KerasClassifier

clf = KerasClassifier(
    model=create_model,
    verbose=0
)
```

```python
# Define search space (these names now match constructor args)
param_dist = {
    "model__neurons": [16, 32, 64],
    "model__hidden_layers": [1, 2, 3],
    "model__activation": ["relu", "tanh"],
    "model__lr": [0.01, 0.001, 0.0001],
    "batch_size": [8, 16, 32],
    "epochs": [30, 50, 80]
}
```

```python
 #pip install -U scikit-learn scikeras
```

```python
# Random Search
from sklearn.model_selection import RandomizedSearchCV

search = RandomizedSearchCV(
    estimator=clf,
    param_distributions=param_dist,
    n_iter=10,
    cv=3,
    verbose=1
)

search_result = search.fit(x_train, y_train_enc)

```

```python
best_clf = search_result.best_estimator_      # SciKeras wrapper
best_model = best_clf.model_                  # actual Keras model

y_pred_tuned = (best_model.predict(x_test) > 0.5).astype(int)
```

```python
# Best results

print("Best Accuracy:", search_result.best_score_)
print("Best Params:", search_result.best_params_)
```

```python
#	Load the dataset 
print("<=========LOADING THE DATASET==========>")
import pandas as pd
import numpy as np
df=pd.read_csv("A3.csv")
df
```

```python
# 	Identify numerical columns in the dataset.
print("<=========NUMERICAL COLUMNS==========>")

df.info()
df.select_dtypes(include='number')
```

```python
# 	Calculate the mean, median, mode, and standard deviation for these columns.
print("<=========DATA DESCRIBE==========>")
df.describe()
```

```python
print("<=========MEAN, MEDIAN, MODE, STANDARD DEVIATION==========>")
print("================MEAN==================")
df.select_dtypes(include=["float64", "int64"]).mean()
#df.me
```

```python
print("================MEDIAN===============")
df.median(numeric_only=True)
```

```python
print("============STANDARD DEVIATION==============")
df.select_dtypes(include="number").std()
```

```python
print("============MODE==========")
df.select_dtypes(include="number").mode().iloc[0]
```

```python
#             HISTOGRAM
df.plot(kind="hist")
```

```python
#           HISTOGRAM
df.hist(figsize=(12,8))
```

```python
#        BOXPLOT
df.plot(kind="box",vert=False)
```

```python
#             Bar Chart
print("Bar Chart")
df.plot.bar(figsize=(8,20),subplots=True)
```

```python
#             BARGRAPH
import matplotlib.pyplot as plt
column_cols=df.select_dtypes(include=["object"]).columns

for col in column_cols:
    df[col].value_counts().plot(kind="bar")
    plt.show()
```
